```python
import kagglehub

# Downloads to Colab's local cache, returns the path
path = kagglehub.dataset_download("mohammedabdeldayem/the-fake-or-real-dataset")
print("Dataset path:", path)
```

```python
import os

for root, dirs, fs in os.walk(path):
    depth = root.replace(path, "").count(os.sep)
    if depth <= 2:
        print(root, "->", dirs[:5], f"({len(fs)} files)")
```

```python
import os

base = os.path.join(path, "for-norm", "for-norm", "training")
print(os.listdir(base))
```

```python
import random

# UPDATE these paths based on Cell 3 output
REAL_DIR = os.path.join(path, "for-norm", "for-norm", "training", "real")
FAKE_DIR = os.path.join(path, "for-norm", "for-norm", "training", "fake")

real_files = [os.path.join(REAL_DIR, f) for f in os.listdir(REAL_DIR) if f.endswith(('.wav', '.mp3'))]
fake_files = [os.path.join(FAKE_DIR, f) for f in os.listdir(FAKE_DIR) if f.endswith(('.wav', '.mp3'))]

print(f"Real: {len(real_files)}, Fake: {len(fake_files)}")

# Subset for time — adjust N based on how it runs
random.seed(42)
N = min(1200, len(real_files), len(fake_files))
real_files = random.sample(real_files, N)
fake_files = random.sample(fake_files, N)
print(f"Using {N} per class")
```

```python
import librosa
import numpy as np

def extract_features(file_path, sr=16000, duration=2.0):
    y, sr = librosa.load(file_path, sr=sr, duration=duration)
    target_len = int(sr * duration)
    if len(y) < target_len:
        y = np.pad(y, (0, target_len - len(y)))
    else:
        y = y[:target_len]

    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
    delta_mfcc = librosa.feature.delta(mfcc)
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    rms = librosa.feature.rms(y=y)

    feats = []
    for f in [mfcc, delta_mfcc, chroma, spec_cent, spec_bw, rolloff, zcr, rms]:
        feats.append(f.mean())
        feats.append(f.std())
        if f.shape[0] > 1:
            feats.extend(f.mean(axis=1))
            feats.extend(f.std(axis=1))
    return np.array(feats, dtype=np.float32)
```

```python
import time

start = time.time()
for fp in real_files[:20]:
    extract_features(fp)
elapsed = time.time() - start
per_file = elapsed / 20
print(f"Per file: {per_file:.3f}s")
print(f"Estimated total for {2*N} files: {per_file * 2*N / 60:.1f} minutes")
```

```python
from joblib import Parallel, delayed
import warnings
warnings.filterwarnings("ignore")  # suppresses the pitch-tuning warning spam

def extract_with_label(fp, label):
    try:
        return extract_features(fp), label
    except Exception as e:
        print("skip", fp, e)
        return None

import time
start = time.time()

tasks = [(fp, 0) for fp in real_files] + [(fp, 1) for fp in fake_files]

results = Parallel(n_jobs=-1, verbose=5)(
    delayed(extract_with_label)(fp, label) for fp, label in tasks
)

results = [r for r in results if r is not None]
X = np.array([r[0] for r in results])
y = np.array([r[1] for r in results])

print(f"Done in {(time.time()-start)/60:.1f} minutes")
print(X.shape, y.shape)
```

```python
np.save("/content/X_features.npy", X)
np.save("/content/y_labels.npy", y)
print("Saved.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import joblib

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Scaling helps some classifiers and doesn't hurt RF
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

clf = RandomForestClassifier(
    n_estimators=300, max_depth=20, random_state=42, n_jobs=-1
)
clf.fit(X_train_s, y_train)

print("Train accuracy:", clf.score(X_train_s, y_train))
print("Test accuracy:", clf.score(X_test_s, y_test))
```

```python
from sklearn.metrics import (
    accuracy_score, f1_score, confusion_matrix,
    roc_curve, classification_report
)
import numpy as np

preds = clf.predict(X_test_s)
probs = clf.predict_proba(X_test_s)[:, 1]

acc = accuracy_score(y_test, preds)
f1 = f1_score(y_test, preds)
cm = confusion_matrix(y_test, preds)

print("Accuracy:", acc)
print("F1:", f1)
print("Confusion Matrix:\n", cm)
print(classification_report(y_test, preds, target_names=["Genuine", "Deepfake"]))

# Per-class accuracy
per_class_acc = cm.diagonal() / cm.sum(axis=1)
print(f"Genuine accuracy: {per_class_acc[0]:.3f}")
print(f"Deepfake accuracy: {per_class_acc[1]:.3f}")

# EER
fpr, tpr, thresholds = roc_curve(y_test, probs)
fnr = 1 - tpr
eer_idx = np.nanargmin(np.abs(fnr - fpr))
eer = (fpr[eer_idx] + fnr[eer_idx]) / 2
print(f"EER: {eer:.4f}")
```

```python
%%writefile /content/predict.py
import sys
import joblib
import numpy as np
from features import extract_features

def predict(file_path):
    clf = joblib.load("model.pkl")
    scaler = joblib.load("scaler.pkl")

    feat = extract_features(file_path).reshape(1, -1)
    feat_scaled = scaler.transform(feat)

    pred = clf.predict(feat_scaled)[0]
    prob = clf.predict_proba(feat_scaled)[0]

    label = "Deepfake (AI-Generated)" if pred == 1 else "Genuine (Human)"
    confidence = prob[pred]

    print(f"Prediction: {label}")
    print(f"Confidence: {confidence*100:.2f}%")
    return label, confidence

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python predict.py <audio_file_path>")
        sys.exit(1)
    predict(sys.argv[1])
```
